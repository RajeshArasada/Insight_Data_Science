```python
df_07.columns = ['SubBorough', '2007_rent', '2007_income', 'SubBorough2', '2007_White', '2007_Black',
                 '2007_Hispanic', 'SubBorough3', '2007_White', '2007_Black', '2007_Hispanic']
```

```python
del df_07['SubBorough2']
del df_07['SubBorough3']
```

```python
df_07.head(3)
```

```python
from sklearn import preprocessing
```

```python
scaled_data = preprocessing.scale(df_07.ix[:,1:])
```

```python
scaled_data
```

```python
df_07.ix[:,1:]
```

```python
data07 = scaled_data
krange = range(1, int(data07.shape[0] / 2))
sum_squares = [cluster.KMeans(n_clusters=k).fit(data07).inertia_ for k in krange]
```

```python
variance_explained = [1.0 - (s / sum(sum_squares)) for s in sum_squares]
plt.figure()
plt.plot(range(1,8), variance_explained)
plt.ylabel('% of explained variance')
plt.xlabel('clusters K')
plt.xlim([1, 7])
```

```python
len(sum_squares), len(variance_explained)
```

```python
plt.figure(figsize=[8, 7])
plt.scatter(data07[:, 0], data07[:, 1],)
plt.title('Neighborhood')
plt.ylabel('income')
plt.xlabel('rent')

## -- 3 Clusters --
clusters_3 = cluster.KMeans(n_clusters=3).fit(data07)
fig, ax = plt.subplots(1, 2, figsize=[9, 7])
fig.suptitle('K Cluster Targets')
ax[0].scatter(data07[:, 0], data07[:, 1], c=clusters_3.predict(data07))

## -- 4 Clusters --
clusters_4 = cluster.KMeans(n_clusters=4).fit(data07)
ax[1].scatter(data07[:, 0], data07[:, 1], c=clusters_4.predict(data07))

fig.savefig('Year2007.png')
```
